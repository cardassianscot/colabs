```python
h = 0.75  # m
v = 3.0   # m/s
g = 9.81  # m/s2
```

Do not alter any of the code below.

```python
#@title The maths and coding bit.
from pylab import *
from matplotlib.ticker import AutoMinorLocator

N = 100    # degree
θmin = 0    # degree
θmax = 75 # degree

θdegs = linspace(θmin, θmax+1, N)
θs = θdegs * pi / 180
ranges = v * cos(θs)/g*(v*sin(θs)+sqrt(v**2*(sin(θs))**2+2*g*h))

plot(θdegs,ranges)
xlabel("θ / degrees")
ylabel("Range / m")
grid(which='major', color='grey')
grid(which='minor', color='silver')
minorticks_on()
show()
```
